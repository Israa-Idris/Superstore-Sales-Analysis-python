**What is the sales trend over time (monthly, yearly)?**

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

Trend for sales

```python
# Convert the "Order Date" column to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')

# Drop rows where "Order Date" could not be converted to a date
df = df.dropna(subset=['Order Date'])

# Group by "Order Date" and sum up sales, then set "Order Date" as the index
monthly_sales = df.groupby('Order Date', as_index=False).sum(numeric_only=True)
monthly_sales = monthly_sales.set_index('Order Date')

# Resample to monthly intervals
monthly_sales = monthly_sales['Sales'].resample('M').sum()

# Plotting
plt.figure(figsize=(25, 8))
plt.plot(monthly_sales)
plt.xlabel("Order Date")
plt.ylabel("Sales")
plt.title("Monthly Sales Trend")
plt.show()
```

```python
# Resample to yearly intervals
yearly_sales = monthly_sales.resample('Y').sum()

# Plotting
plt.figure(figsize=(25, 8))
plt.plot(yearly_sales, label='Yearly Sales', marker='o')
plt.xlabel("Order Date")
plt.ylabel("Sales")
plt.title("Yearly Sales Trend")
plt.legend()
plt.show()
```

The general trend of the company is sales increse , but there are some periods in which profits decrease and then rise again. I think it's because of the discounts




Trend for Profit

```python
# Ensure Order Date is in datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Group by Order Date and sum only numeric columns
monthly_sales = df.groupby('Order Date')[['Profit']].sum()

# Resample to monthly intervals
monthly_sales = monthly_sales.resample('M').sum()

# Plot
plt.figure(figsize=(25, 8))
plt.plot(monthly_sales['Profit'], label='Monthly Profit', marker='o')
plt.xlabel("Order Date")
plt.ylabel("Profit")
plt.title("Monthly Profit Trend")
plt.legend()
plt.show()
```

```python
# Resample to yearly intervals
yearly_sales = monthly_sales.resample('Y').sum()

# Plotting
plt.figure(figsize=(25, 8))
plt.plot(yearly_sales, label='Yearly profit', marker='o')
plt.xlabel("Order Date")
plt.ylabel("profit")
plt.title("Yearly profit Trend")
plt.legend()
plt.show()
```

The general trend of the company is profit increase

What is the average profit margin for each product category?

```python
# Group the data by product category and calculate the average profit for each category
avg_profit_margin_by_category = df.groupby('Category')['Profit'].mean()

print(avg_profit_margin_by_category)
```

```python
df['Profit Margin'] = df['Profit'] / df['Sales']

# Group the data by product category and calculate the average profit margin for each category
avg_profit_margin_by_category = df.groupby('Category')['Profit Margin'].mean()

# Plot the average profit margin for each category as a bar chart
avg_profit_margin_by_category.plot(kind='bar')

# Add a title and labels to the chart
plt.title("Average Profit Margin by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Average Profit Margin")

plt.show()
```
